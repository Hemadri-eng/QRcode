```python
import qrcode 
import image
```

```python
qr = qrcode.QRCode(version = 15,
                   box_size=5,
                   border = 10)
```

```python
data = "https://www.youtube.com/watch?v=rNXmANj72ps&list=RDrNXmANj72ps&start_radio=1"
```

```python
qr.add_data(data)
```

```python
qr.make(fit = True)
```

```python
img = qr.make_image(fill = "black",
                   back_color = "white" )
```

```python
img.save("test.png")
```
